```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

def duration(Signal, Fs):
    Ts = 1/Fs
    sig_duration = len(Signal)/Fs
    t = np.arange(0, sig_duration, Ts)
    return sig_duration, t

data1 = pd.read_csv("Downloads\\sens1.csv")
A1 = data1[['ACCELEROMETER X (m/s²)','ACCELEROMETER Y (m/s²)','ACCELEROMETER Z (m/s²)']]
G1 = data1[['GYROSCOPE X (rad/s)', 'GYROSCOPE Y (rad/s)', 'GYROSCOPE Z (rad/s)']]
data1.to_csv('rest1.csv')
data2 = pd.read_csv("Downloads\\sens2.csv")
A2 = data2[['ACCELEROMETER X (m/s²)','ACCELEROMETER Y (m/s²)','ACCELEROMETER Z (m/s²)']]
G2 = data2[['GYROSCOPE X (rad/s)', 'GYROSCOPE Y (rad/s)', 'GYROSCOPE Z (rad/s)']]
data2.to_csv('walk1.csv')
data3 = pd.read_csv("Downloads\\sens3.csv")
A3 = data3[['ACCELEROMETER X (m/s²)','ACCELEROMETER Y (m/s²)','ACCELEROMETER Z (m/s²)']]
G3 = data3[['GYROSCOPE X (rad/s)', 'GYROSCOPE Y (rad/s)', 'GYROSCOPE Z (rad/s)']] 
data3.to_csv('run1.csv')

fig, axes=plt.subplots(nrows=3,ncols=2,figsize = (10,7))
restA,restG,walkA,walkG,runA,runG = axes.flatten()

restA.plot(A1)
restA.set_xlabel('Time(s)',fontsize=10)
restA.set_ylabel('Acceleration m/s^2 (rest)',fontsize=10)

walkA.plot(A2)
walkA.set_xlabel('Time(s)',fontsize=10)
walkA.set_ylabel('Acceleration m/s^2 (walk)',fontsize=10)

runA.plot(A3)
runA.set_xlabel('Time(s)',fontsize=10)
runA.set_ylabel('Acceleration m/s^2 (run)',fontsize=10)

restG.plot(G1)
restG.set_xlabel('Time(s)',fontsize=10)
restG.set_ylabel('Gyroscope rad/s (rest)',fontsize=10)

walkG.plot(G2)
walkG.set_xlabel('Time(s)',fontsize=10)
walkG.set_ylabel('Gyroscope rad/s (walk)',fontsize=10)

runG.plot(G3)
runG.set_xlabel('Time(s)',fontsize=10)
runG.set_ylabel('Gyroscope rad/s (run)',fontsize=10)
fig.tight_layout()

print ("Тривалість сигналу:")
duration(data1, 20)
```
