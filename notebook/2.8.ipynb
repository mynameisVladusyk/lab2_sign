```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Зчитування сигналу
data = np.loadtxt("Downloads\\laba\\TBI_ICP.txt")
np.savetxt("intracranial pressure",data)
Fs= 125
T = len(data)/Fs
t = np.linspace(0,T,round(Fs*T))
#Побудова графіка
fig, txt = plt.subplots(figsize=(12,7))
txt.set_xlabel('time,s')
txt.set_ylabel('mmHg')
txt.set_title('Data from text file TBI_ICP.txt')
txt.plot(t,data)
plt.show()
```

```python
def duration(Signal, Fs):
    Ts = 1/Fs
    sig_duration = len(Signal)/Fs
    t = np.arange(0, sig_duration, Ts)
    return sig_duration, t
duration(data, Fs)
```
