```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import sys 
#Функція побудови графіка
def plot(signal, Fs, Tstart, Tstop):
    if Tstart <0 or Tstop > len(signal) or Tstart > Tstop:
        sys.exit('Check your input data')
    else:
        fig = plt.figure()
        t = np.linspace(Tstart, Tstop, len(signal))
        plt.plot(t, signal)
        plt.show()
        return signal, Tstart, Tstop
#Зчитування сигналу
signal_ecg = np.load("Downloads\\lab2\\norm\\1600714586.npz")
fs = signal_ecg['fs']
sig1 = signal_ecg['signal']
#Побудова сигналу
plot(sig1, fs, 20,23)
```

```python
def duration(Signal, Fs):
    Ts = 1/Fs
    sig_duration = len(Signal)/Fs
    t = np.arange(0, sig_duration, Ts)
    return sig_duration, t
duration(sig1,fs)
```
