```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Функція побудови графіка
def plot(x,y,n,title,ylabel):
    plt.subplot(2,1,n)
    plt.plot(x,y)
    plt.ylabel(ylabel)
    plt.xlabel('сек')
    plt.title(title)
    plt.grid()
#Зчитування сигналу 
data = pd.read_csv('Downloads\\laba\\Subject1_SpO2Hr.csv', sep = ',')
t = data["Elapsed time(seconds)"]
saturation = data["SpO2(%)"]
heartbeat = data["hr (bpm)"]

saturation_aver = []
heartbeat_aver = []
interval = 30
#Визначення середніх значень
for i in range(0,len(t),interval):
    if (i+interval) > len(t):
        saturation_aver.append(np.mean(saturation[i:]))
        heartbeat_aver.append(np.mean(heartbeat[i:]))
    else:
        saturation_aver.append(np.mean(saturation[i:i+interval]))
        heartbeat_aver.append(np.mean(heartbeat[i:i+interval]))
#Побудова графіків
plot(t,saturation,1,"Сатурація артеріальної крові киснем", "Sp02(%)")
plt.show()

plot(t,heartbeat,1,"Значення серцевого ритму", "hr (bpm)")
plt.show()

plot(np.linspace(0,t.T.values[-1], len(saturation_aver)),saturation_aver, 1, "Усереднена сатурація артеріальної крові киснем", "SpO2(%)")
plt.show()

plot(np.linspace(0,t.T.values[-1], len(heartbeat_aver)),heartbeat_aver, 2, "Усереднена значення серцевого ритму", "hr (bpm)")
plt.show()
```

```python
def duration(Signal, Fs):
    Ts = 1/Fs
    sig_duration = len(Signal)/Fs
    t = np.arange(0, sig_duration, Ts)
    return sig_duration, t
duration(data,256)
```
