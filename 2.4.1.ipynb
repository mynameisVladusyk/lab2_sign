```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
# 
signal_ecg = np.load("Downloads\\lab2\\norm\\1600714586.npz")
fs = signal_ecg['fs']
units = signal_ecg['units']
signal = signal_ecg['signal']
labels = signal_ecg['labels']
labels_indexes = signal_ecg['labels_indexes']

y_labels = [signal[i] for i in labels_indexes]
# 
plt.plot(signal)

plt.title('Healthy person')
plt.xlabel('time(s)')
plt.ylabel('U,mV')

plt.scatter(labels_indexes, y_labels, color = 'aqua', s = 40, marker = 'o', label = ' ')
plt.legend()
plt.show()
```

```python
# 
def duration(Signal, Fs):
    Ts = 1/Fs
    sig_duration = len(Signal)/Fs
    t = np.arange(0, sig_duration, Ts)
    return sig_duration, t
duration(signal, 250)
```
